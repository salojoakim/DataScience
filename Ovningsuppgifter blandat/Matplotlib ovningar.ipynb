```python
import matplotlib.pyplot as plt
import numpy as np

# Create all possible outcomes of rolling two dice
die1 = np.arange(1, 7)  # Values for the first die (1-6)
die2 = np.arange(1, 7)  # Values for the second die (1-6)

# Create a grid of possible outcomes
outcomes = [(d1, d2) for d1 in die1 for d2 in die2]

# Separate the outcomes into two lists (for plotting)
x_values = [outcome[0] for outcome in outcomes]  # Values of die 1 (x-axis)
y_values = [outcome[1] for outcome in outcomes]  # Values of die 2 (y-axis)

# Create the plot
plt.figure(figsize=(8, 8))

# Plot all the outcomes, using different colors for the two dice
plt.scatter(x_values, y_values, c='blue', label="Die 1 (x-axis)", s=100, edgecolor='black')
plt.scatter(y_values, x_values, c='red', label="Die 2 (y-axis)", s=100, edgecolor='black')

# Labeling the axes
plt.xlabel("Die 1")
plt.ylabel("Die 2")

# Add a title and legend
plt.title("All Possible Outcomes of Rolling Two Dice")
plt.legend()

# Show a grid for better visibility
plt.grid(True)

# Show the plot
plt.show()
```
